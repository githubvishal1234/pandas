```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,how="outer")   #takes all key values
```

```python
pd.merge(df1,df2,how="cross")
```

```python
pd.merge(df1,df2,how="left",left_on="key2",right_on="key5")
```
