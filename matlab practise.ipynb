```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.get_dataset_names()
```

```python
sns.histplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=100)
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
sns.barplot(x=tips.total_bill[0:5],y=tips.tip[0:5])
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue="smoker")
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style="size",hue="smoker")
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style="time",hue="smoker")
```
