```python
plt.figure()
plt.subplot(2,2,1)  #2 rows 2 columns
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)  # 2 rows one column
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x),color="blue")
plt.plot(x,np.cos(x),color="green")
plt.plot(x,np.cos(x+1),color="cyan")
```

```python
plt.plot(x,np.sin(x))
plt.ylim(-1.5,1.5)
plt.xlim(0,20)
plt.axis("equal")
plt.grid()
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.ylim(-1.5,1.5)
plt.xlim(0,20)
plt.axis("equal")
plt.grid()
plt.title("This is sin axis")
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.ylim(-1.5,1.5)
plt.xlim(0,20)
plt.axis("equal")
plt.grid()
plt.title("This is sin axis")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x,np.sin(x),color="red",label="sinx")
plt.plot(x,np.cos(x),color="blue",label="cosx")
plt.ylim(-1.5,1.5)
plt.xlim(0,20)
plt.axis("equal")
plt.grid()
plt.title("This is sin graph")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
langs=["java","kotlin","android","python","datascience"]
students=[10,20,5,50,15]
sum=0
for i in students:
    sum+=i
print(sum)
plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.show()
```

```python
my_explode=(0.0,0.1,0.0,0.1,0.0)
langs=["java","kotlin","android","python","datascience"]
students=[10,20,5,50,15]
sum=0
for i in students:
    sum+=i
print(sum)
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=my_explode,shadow=True)
plt.show()
```
